```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
iris=load_iris()
```

```python
iris_df=pd.DataFrame(data=iris.data,columns=iris.feature_names)
```

```python
iris_df
```

```python
scaler=StandardScaler()
```

```python
data=scaler.fit_transform(iris_df)
```

```python
data
```

```python
plt.scatter(data[:,0],data[:,1],cmap="viridis")
```

```python
pca=PCA()
```

```python
pca_features=pca.fit_transform(data)
```

```python
plt.figure(figsize=(10,6))
plt.plot(pca.explained_variance_ratio_,marker="o",linestyle="--")      ###np.cumsum(pca)
plt.show()
```

```python
pca_features
```

```python
pca=PCA(n_components=2)
```

```python
pca_features=pca.fit_transform(iris_df)
```

```python
plt.figure(figsize=(10,6))
plt.plot(pca.explained_variance_ratio_,marker="o",linestyle="--")      ###np.cumsum(pca)
plt.show()
```

```python
pca_features
```

```python
inertia=[]
for i in range(1,11):
    kmeans= KMeans(n_clusters=i,random_state=42)
    kmeans.fit(pca_features)
    inertia.append(kmeans.inertia_)
    
```

```python
inertia
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,11),inertia,marker="o",linestyle="--")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS Valkye")
plt.title("Elbow Graph")
```

```python
kmeans=KMeans(n_clusters=4)
kmeans.fit(pca_features)
```

```python
clusters_labels=kmeans.labels_
```

```python
clusters_labels
```
